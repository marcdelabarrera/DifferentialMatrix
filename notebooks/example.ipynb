# 1D test

```julia
function f(x)
    return sin.(x)
end

function f_prime(x)
    return cos.(x)
end
```

```julia
Delta = 0.01
x = collect(0:Delta:10)
plot(x, f(x), label = "f(x)")
plot!(x, f.(x), label = "f(x)", seriestype = :scatter, markersize = 1, color = :blue)
plot!(x, f_prime(x), label="f'(x)")
plot!(x, create_diff_matrix(size(x), Delta = Delta, direction="forward")*f(x), label="f'(x) forward")
plot!(x, create_diff_matrix(size(x), Delta = Delta, direction="backward")*f(x), label="f'(x) backward")
```

```julia
Delta = 0.02
x = exp.(0:Delta:3)

plot(x, f(x), label = "f(x)")
plot!(x, f.(x), label = "f(x)", seriestype = :scatter, markersize = 1, color = :blue)
plot!(x, f_prime(x), label="f'(x)")
plot!(x, create_diff_matrix(size(x), Delta = diff(x), direction="forward")*f(x), label="f'(x) forward")
plot!(x, create_diff_matrix(size(x), Delta = diff(x), direction="backward")*f(x), label="f'(x) backward")
```
